```python
import pandas as pd

# Ensure 'Date' is datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract Month and Year
df['Month'] = df['Date'].dt.month_name()    # e.g., January
df['Year'] = df['Date'].dt.year             # e.g., 2025
df['Day_Name'] = df['Date'].dt.day_name()   # e.g., Monday

# Filter for a specific month and year
month_name = "January"
year_value = 2025

month_year_data = df[(df['Month'] == month_name) & (df['Year'] == year_value)]

# Sort by AQI descending (worst air quality first)
month_year_data = month_year_data.sort_values(by='AQI', ascending=False)

# Select relevant columns
month_year_data = month_year_data[['City', 'Date', 'Month', 'Year', 'Day_Name', 'AQI', 'AQI_Category']]

# Display the result
month_year_data
```

```python
import pandas as pd

# Ensure 'Date' column is datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract Year, Month, and Day for convenience
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day
df['Day_Name'] = df['Date'].dt.day_name()

# Filter for New Year's Day (January 1) across all years
new_year_data = df[(df['Month'] == 'January') & (df['Day'] == 1)]

# Select relevant columns
new_year_data = new_year_data[['City', 'Year',  'AQI', 'AQI_Category']]

# Sort by Year for better readability
new_year_data = new_year_data.sort_values(by='Year',ascending=False)

# Display the result
new_year_data
```

```python
df['Temperature_Range'] = df['Temperature_Max (°C)'] - df['Temperature_Min (°C)']
df['Is_Rainy'] = df['Rainfall (mm)'] > 0
```

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Weekday'] = df['Date'].dt.day_name()
```

```python
aqi_map = {
    'Good': 1,
    'Satisfactory': 2,
    'Moderate': 3,
    'Poor': 4,
    'Very Poor': 5,
    'Severe': 6
}

df['AQI_Level_Code'] = df['AQI_Category'].map(aqi_map)
```

```python
df[df['AQI'] > 200]
df[df['City'] == 'Delhi']
df[(df['Rainfall (mm)'] > 5) & (df['Humidity (%)'] > 70)]

df[['City', 'AQI', 'Temperature_Avg (°C)']]
```

```python
df.sort_values(by='AQI', ascending=False)
df['AQI_Rank'] = df['AQI'].rank(ascending=False)
```
